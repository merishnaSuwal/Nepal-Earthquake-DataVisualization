## Get the data for no. of persons injured for Bagmati Zone

```python
viz_zone = data_input[data_input['Zone']=='Bagmati']
```

### Visualize the data [Number of persons injured, Districts]

```python
# viz_injured_data.plot(x='District', y='In Number', kind='bar')
plt.figure(figsize = (28,10))
plt.title('Number of persons injured in Bagmati zone',fontsize=25)
sns.set_context('poster',font_scale=2,rc={"lines.linewidth": 2})
sns_plot = sns.barplot(x='District',y='In Number',data=viz_zone,ci=None)
plt.ylim(0,300000)

# Create names
# plt.xticks(y_pos, bars)
plt.xlabel('Districts',fontsize=25)
plt.ylabel('Injured people',fontsize=25)
plt.tight_layout()
```

```python
figure = sns_plot.get_figure()
figure.savefig('Plots/Bagmati zone plot.png', dpi=300)
```

## Visualizing data for zones in Central Development region

```python
viz_devregion = data_input[data_input['Development Region']=='Central']
```

```python
viz_devregion['Zone'].unique()
```

### Number of people injured zone-wise due to 2015 earthquake

```python
g = sns.factorplot(x="District", y="In Number",
                    col="Zone",ci= None,col_wrap=3,
                   data=viz_devregion, kind="bar",
                   size=9, aspect=1.5);
g.set_axis_labels("Districts", "Number of people injured")
g.set(ylim=(0, 300000))
g.set_titles("{col_name} {col_var}")
g.set_xticklabels(rotation=90)
```

```python
fig = g.get_figure()
fig.savefig('Plots/Central Development region.png', dpi=300)
```
